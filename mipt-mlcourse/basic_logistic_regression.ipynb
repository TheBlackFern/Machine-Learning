# Logistic regression

```python
#import packages

from sklearn.linear_model import LogisticRegression      # import LogisticRegression classifier
from matplotlib import pyplot as plt                     # package for visualisations
import numpy as np                                       # package for linear algebra computations
```

```python
# Let us set X and y

X1 = np.random.normal(0, 1.5, (100,1))                     # first set 
X2 = np.random.normal(3, 1.5, (100,1))                 # second set
X3 = np.random.normal(7, 3, (100,1))                 # third set
Y1 = np.random.normal(3, 1.5, (100,2))
Y2 = np.random.normal(0, 1.5, (100,2))
Y3 = np.random.normal(7, 3, (100,2))
Y1[:,1] = 0                                             # labels for the first set
Y2[:,1] = 1                                              # labels for the second set
Y3[:,1] = 2                                              # labels for the third set
R1 = np.hstack([X1,Y1])
R2 = np.hstack([X2,Y2])
R3 = np.hstack([X3,Y3])

S = np.vstack([R1,R2,R3])                                   # stack to sets
X = S[:,:2]                                              # extract X
y = S[:,2]                                               # extract y
# Visualize a set of pairs of (x,y) on the plane

plt.figure(figsize=(6,4))

plt.scatter(X[:,0],X[:,1], c=y)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.title('Figure 1: The scater plot for data in the plane of features \n');
```

**Task:** build a classification model for the given data X and y, using *LogisticRegression* from *sklearn*.

### Logistic Regression Classifier

```python
# your code here
clf = LogisticRegression().fit(X,y)
y_hat = clf.predict(X)
# Probability estimates

P = clf.predict_proba(X)
# Let us compute mean accuracy on the given test data and labels

clf.score(X, y), y_hat
# Visualize a set of pairs of (x,y) on the plane

from sklearn.inspection import DecisionBoundaryDisplay

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
DecisionBoundaryDisplay.from_estimator(
    clf, X, response_method="predict", cmap=plt.cm.Paired, ax=axes[1]
)

axes[1].scatter(X[:,0],X[:,1], c=y_hat)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

axes[0].scatter(X[:,0],X[:,1], c=y)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

fig.tight_layout()
```
